# Scaling and Normalization Techniques

Scaling and Normalization are data preprocessing techniques used to bring numerical features to a common scale.

Different features in a dataset may have different ranges. For example, Sales may range from 0 to 2000, while Order Item Quantity may range from 1 to 5.

If the data is not scaled, features with larger values may dominate machine learning algorithms. Therefore, scaling and normalization are performed to improve model performance and training efficiency.

## Common Scaling Techniques

1. Standard Scaling (Z-Score Scaling)
   - Transforms data to have mean = 0 and standard deviation = 1.
   - Formula:
     Z = (X - Mean) / Standard Deviation

2. Min-Max Scaling (Normalization)
   - Rescales values between 0 and 1.
   - Formula:
     X_scaled = (X - Xmin) / (Xmax - Xmin)

3. Robust Scaling
   - Uses median and interquartile range.
   - Suitable for datasets with outliers.

4. Max Absolute Scaling
   - Scales values between -1 and 1.
   - Preserves sparsity in the data.

## Normalization

Normalization is a preprocessing technique that rescales numerical features to a fixed range, typically between 0 and 1.

Formula:

X_normalized = (X - Xmin) / (Xmax - Xmin)

Advantages:
- Removes differences in scale.
- Improves model convergence.
- Useful for distance-based algorithms such as KNN and Neural Networks.

# Encoding and Its Techniques

Encoding is the process of converting categorical data into numerical format so that machine learning algorithms can process the data effectively.

The following encoding techniques are commonly used in machine learning.

## 1. Label Encoding

Label Encoding assigns a unique integer value to each category.

Example:

Red   → 0
Blue  → 1
Green → 2

Advantages:
- Simple and easy to implement.
- Requires less memory.
- Suitable for tree-based algorithms.

Disadvantages:
- May introduce an artificial order among categories.

## 2. One-Hot Encoding

One-Hot Encoding creates separate binary columns for each category.

Example:

Color

Red
Blue
Green

becomes

Color_Red   Color_Blue   Color_Green
1           0            0
0           1            0
0           0            1

Advantages:
- No ordinal relationship is created.
- Suitable for nominal categorical variables.

Disadvantages:
- Increases the number of columns.

## 3. Binary Encoding

Binary Encoding converts categories into numerical values and then into binary format.

Example:

A → 1 → 001
B → 2 → 010
C → 3 → 011

Advantages:
- Reduces dimensionality compared to One-Hot Encoding.
- Efficient for high-cardinality features.

Disadvantages:
- More difficult to interpret.

## 4. Frequency Encoding

Frequency Encoding replaces each category with its frequency of occurrence.

Example:

Chennai → 500
Trichy  → 300
Madurai → 150

Advantages:
- Handles high-cardinality features efficiently.
- Requires less memory.

Disadvantages:
- Categories with the same frequency receive the same value.

# Feature Selection

Feature Selection is the process of selecting the most relevant features from the dataset while removing redundant or less important attributes.

Benefits:
- Reduces model complexity.
- Improves training speed.
- Reduces overfitting.
- Improves model performance.

## Methods of Feature Selection

Feature Selection techniques are broadly classified into:

### 1. Filter Methods
Features are selected using statistical measures before model training.

Examples:
- Correlation Analysis
- Chi-Square Test
- ANOVA Test
- Information Gain

#### Advantages
- Fast and computationally efficient
- Suitable for large datasets
- Independent of machine learning algorithms

#### Disadvantages
- Ignores feature interactions
- May select redundant features

### 2. Wrapper Methods

Wrapper methods evaluate subsets of features by repeatedly training a machine learning model.

Examples:
- Forward Selection
- Backward Elimination
- Recursive Feature Elimination (RFE)

#### Advantages
- High accuracy
- Considers feature interactions
- Produces optimal feature subsets

#### Disadvantages
- Computationally expensive
- Time consuming for large datasets
- Risk of overfitting

### 3. Embedded Methods

Feature selection is performed during model training.

Examples:
- Lasso Regression
- Random Forest Feature Importance
- XGBoost Feature Importance

#### Advantages
- Faster than wrapper methods
- Considers feature interactions
- Good predictive performance

#### Disadvantages
- Model dependent
- Less flexible compared to filter methods

## Logistic Regression


Logistic Regression was applied to predict the Late_delivery_risk target variable. The model was trained using the preprocessed dataset after handling missing values, encoding categorical variables, and normalization. The model achieved a training accuracy of 99.33% and a testing accuracy of 99.33%, indicating strong predictive performance. The confusion matrix showed only 242 misclassifications out of 36,104 test samples, demonstrating the model's effectiveness in identifying delivery risks.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create model
lr = LogisticRegression(
    max_iter=5000,
    solver='liblinear'
)

# Train model
lr.fit(X_train, y_train)

# Predictions
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)

# Accuracy
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print("Training Accuracy:", train_acc)
print("Testing Accuracy:", test_acc)

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_test_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```

## Gaussian Naive Bayes

It was selected as an additional classification model to predict Late_delivery_risk.

Gaussian Naive Bayes is a probability-based supervised machine learning algorithm that applies Bayes' Theorem and assumes that all features are independent of each other. It is called "Gaussian" because it assumes that numerical features follow a normal (Gaussian) distribution.

The model calculates the probability of each class and assigns the class with the highest probability as the prediction.

Advantages:
• Fast training and prediction
• Works well on large datasets
• Probability-based classification
• Simple and easy to implement

Disadvantages:
• Assumes feature independence
• Performance may decrease when features are highly correlated

The model performance is evaluated using Accuracy Score, Classification Report, and Confusion Matrix.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create Gaussian Naive Bayes model
gnb = GaussianNB()

# Train the model
gnb.fit(X_train, y_train)

# Predictions
y_train_pred = gnb.predict(X_train)
y_test_pred = gnb.predict(X_test)

# Accuracy
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print("Training Accuracy:", train_acc)
print("Testing Accuracy:", test_acc)

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_test_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```

The results show that Gaussian Naive Bayes is an effective probability-based classification algorithm for predicting late delivery risk. ## Hyperparameter Tuning
